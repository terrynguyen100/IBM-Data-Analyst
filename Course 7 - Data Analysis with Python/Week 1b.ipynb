```python
import pandas as pd
import numpy as np

df = pd.read_excel('adult.xlsx')
df.head(2)
```

```python
# replacing
df.replace('?', 'Not Indicated', inplace = True)
```

```python
#1 How many men and women (sex feature) are represented in this dataset?
df['gender'].value_counts()
```

```python
#2 What is the average age (age feature) of women?
dftemp = df.loc[df['gender'] == 'Female']
dftemp['age'].mean()
```

```python
#3 What is the percentage of German citizens (native-country feature)?
dftemp = df.loc[df['native-country'] == 'Germany']
percent = len(dftemp) / len(df) * 100 
print("{:.2f}%".format(percent))
```

```python
#4 What are the mean and standard deviation of age for those who earn more than 50K per year 
# (salary feature) and those who earn less than 50K per year?
df.rename(columns ={'income' : 'salary'}, inplace = True)

dftemp = df.loc[df['salary'] == '>50K']
print('Mean of age of the >50K group: ', dftemp['age'].mean())
print('STD of age of the >50K group: ', dftemp['age'].std())

dftemp = df.loc[df['salary'] == '<=50K']
print('Mean of age of the <=50K group: ', dftemp['age'].mean())
print('STD of age of the <=50K group: ', dftemp['age'].std())
```

```python
#5 Is it true that people who earn more than 50K have at least a high school education? 
# (education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters or Doctorate feature)
dftemp = df.loc[df['salary'] == '>50K']
dftemp['education'].value_counts()

# Conclusion: No, some with more than 50K do not have at least a high school education 
```

```python
#6 Display age statistics for each race (race feature) and each gender (sex feature). Use groupby() and describe(). 
# Find the maximum age of men of Amer-Indian-Eskimo race.

# first approach, we groupby race and gender, and select on column age and get its statistics, all in 1 step and put in a multiindex dataframe
dftemp = df.groupby(['race', 'gender'])['age'].describe()
dftemp
```

```python
dftemp.loc['Amer-Indian-Eskimo' ,'Male']['max']
# Notice how we do not use max() because the dftemp is now a dataframe of the statistics
```

```python
#6b different way of solving the #6 question. We pull only the max age into the dataframe with multiindex
dftemp = df.groupby(['race', 'gender'])['age'].max()
dftemp
```

```python
dftemp.loc['Amer-Indian-Eskimo', 'Male']
```

```python
#7 Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (marital-status feature)?
# Consider as married those who have a marital-status starting with Married (Married-civ-spouse, 
# Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.

total_men_50k = df.loc[(df['gender'] == 'Male') & (df['salary'] == '>50K')]
married_men_50k = df.loc[(df['gender'] == 'Male') & (df['salary'] == '>50K') & (df['marital-status'].str.startswith('Married'))]

print(len(married_men_50k))
print(len(total_men_50k))
percent = len(married_men_50k) / len(total_men_50k) * 100
print("Propotion of married men making more than 50K: {:.2f}%".format(percent))
```

```python
#8  What is the maximum number of hours a person works per week (hours-per-week feature)?
# How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?

dftemp = df['hours-per-week'].value_counts().to_frame(name = 'Counts')
dftemp['Percentage'] = (dftemp['Counts'] / dftemp['Counts'].sum() * 100)
dftemp.index.names = ['Hour-per-week']
dftemp
```

```python
#9a Count the average time of work (hours-per-week) for those who earn a little and a lot (salary) for each country (native-country). 

dftemp = df.groupby(['native-country', 'salary'])['hours-per-week'].mean().to_frame()
dftemp.head()
```
